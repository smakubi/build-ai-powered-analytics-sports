```python
dbutils.widgets.text(
    "CATALOG",
    "",
    "Catalog Name"
)
CATALOG = dbutils.widgets.get("CATALOG")
```

```python
schemas_df = spark.sql(
    f"SHOW SCHEMAS IN {CATALOG}"
)
```

```python
system_schemas = [
    "builtin",
    "session",
    "information_schema"
]

for row in schemas_df.collect():
    schema_name = row['databaseName'].strip()
    if (
        schema_name in system_schemas
        or schema_name.startswith("SYS")
        or schema_name.startswith("DATABRICKS")
    ):
        continue
    spark.sql(
        f"DROP SCHEMA IF EXISTS `{CATALOG}`.`{schema_name}` CASCADE"
    )

spark.sql(
    f"DROP CATALOG `{CATALOG}`"
)
```
