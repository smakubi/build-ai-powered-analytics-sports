Uses the `COPY INTO` SQL command to load raw game data from Parquet files into the bronze table. The command merges schemas if needed and ensures all game data is available for further processing and analysis.

```python
spark.sql(
    f"""
    COPY INTO {BRONZE_GAMES_TABLE}
    FROM '{GAMES_DIRECTORY}'
    FILEFORMAT = PARQUET
    PATTERN = '*.parquet'
    FORMAT_OPTIONS ('mergeSchema' = 'true')
    COPY_OPTIONS (
        'mergeSchema' = 'true', 
        'force' = 'false');
    """
)
```
